### Create Utilization Rate Features

To better understand resource usage in hospitals, we calculated utilization rates for both adult and COVID-specific bed usage:

- `inpatient_bed_covid_utilization_rate`
- `adult_icu_bed_utilization_rate`
- `adult_icu_bed_covid_utilization_rate`

Each rate was computed as:

```python
utilization_rate = numerator / denominator

````python
# Create utilization rates
df['inpatient_bed_covid_utilization_rate'] = df['inpatient_bed_covid_utilization_numerator'] / df['inpatient_bed_covid_utilization_denominator']
df['adult_icu_bed_utilization_rate'] = df['adult_icu_bed_utilization_numerator'] / df['adult_icu_bed_utilization_denominator']
df['adult_icu_bed_covid_utilization_rate'] = df['adult_icu_bed_covid_utilization_numerator'] / df['adult_icu_bed_covid_utilization_denominator']
````

````python
# Drop the original columns used to create utilization rates 
df.drop(columns=[
    'inpatient_bed_covid_utilization_numerator',
    'inpatient_bed_covid_utilization_denominator',
    'adult_icu_bed_utilization_numerator',
    'adult_icu_bed_utilization_denominator',
    'adult_icu_bed_covid_utilization_numerator',
    'adult_icu_bed_covid_utilization_denominator'
], inplace=True)
````
